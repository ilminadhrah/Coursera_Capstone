#step 1 : scrape data from wiki

```python
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

```python
headers = ["Postal Code", "Borough", "Neighbourhood"]
```

```python
df['Borough'].value_counts()
```

```python
# step 2 data cleaning
#2.1 Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

import numpy as np

df['Borough'].replace("Not assigned", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
#drop not assigned/nan on borough

df.dropna(subset=["Borough"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
#2.2 Combine borough according to postal code 

df_grouped = df.groupby(['Postal Code','Borough'], as_index=False).agg(lambda x: ','.join(x))

df_grouped.head()
```

```python
#2.3 combine neighbourhood, seperated by comma

mask = df_grouped["Neighbourhood"] == "Not assigned"
df_grouped.loc[mask, "Neighbourhood"] = df_grouped.loc[mask, "Borough"]
```
